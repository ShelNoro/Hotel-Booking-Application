# 2. Data Cleaning

Check for Duplicates

```python
booking[booking.duplicated()]
```

Total number of Duplicate rows(repeated entries)

```python
booking[booking.duplicated()].shape[0]
```

Remove Duplicate rows(repeated entries)

```python
booking.drop_duplicates(inplace=True)
```

```python
booking.shape
```

Get count of Null values per coulmn

```python
booking.isna().sum()
```

The Columns having Null values

```python
lit= np.array(booking.columns[np.where(booking.isna().sum()!=0)])
lit
```

```python
for i in lit:
    print(i,": ",booking[i].value_counts(),"\n")
```

Since the columns Company Agent and Children are Float valueswe can Replace NAN with 0.0

```python
booking[["children", "company", "agent"]]=booking[["children", "company", "agent"]].fillna(0.0)
```

The column Country is a object data type, hence we can replace NAN values with Unknown value

```python
booking["country"] = booking["country"].fillna("unknown")
```

Check if any Null Values are left

```python
null_values= np.array(booking.columns[np.where(booking.isna().sum()!=0,)])
null_values
```

# 3. Creating Required Columns

Creating a column for Date

```python
from datetime import date,datetime
booking['arrival_date']=pd.to_datetime(booking['arrival_date_year'].astype(str)+'-'+booking['arrival_date_month']+'-'+booking['arrival_date_day_of_month'].astype(str),format='%Y-%B-%d')
```

To get the total number of guests

```python
booking['Total_guests']= booking['adults']+booking['children']+booking['babies']
```

# 4. Conveting Data types of columns

```python
booking.info()
```

The column children is in 'float' data type and needs to be in 'int'

```python
booking["children"]=booking["children"].astype(int)
```

Convert column "reservation_status_date" into Datetime datatype

```python
booking['reservation_status_date'] = pd.to_datetime(booking['reservation_status_date'])
```

Converting column 'Total Guests' from 'float' to 'int'

```python
booking['Total_guests']=booking['Total_guests'].astype(int)
```

5.	What is the cancellation rate for each market segment?

1.	How does the lead time affect the likelihood of booking cancellations?

```python
likelihood=booking[['is_canceled','lead_time']]
likelihood.sort_values(by='lead_time', ascending=False)
```

```python
new=[]
likelihood=pd.DataFrame(likelihood)
bin = [0,100,200,300,400,500,600,700,800]
lables=['0-100','100-200','200-300','300-400','400-500','500-600','600-700','700-800']
likelihood['range']=pd.cut(likelihood['lead_time'],bins=bin,labels=lables)
likelihood
```

```python
likelihood_range=likelihood.groupby(['is_canceled','range'])['lead_time'].value_counts()
likely=likelihood_range.dropna().round(2).reset_index()
likely=likely.groupby(['is_canceled','range'])['count'].sum().reset_index()
print(likely)
sns.lineplot(data=likely[likely['is_canceled']==0], x='range', y='count', label ='No Cancellations',color='blue')
sns.lineplot(data=likely[likely['is_canceled']==1], x='range', y='count', label ='Cancellations',color='red')
plt.xticks(rotation=60)
plt.show()
```
